For example, let's say we also want to tell the person what **generation** they are a part of. Now generation boundaries are widely debated and felxible but they are roughly:

So, let's say we want to tell people what generation they are as well as how many years ago they were born. Our first version might look something like this: 

````python
current_year = 2023
birth_year = input("What year were you born? ")
birth_year = int(birth_year)

if 1901 <= birth_year <= 1924:
  generation = "The Greatest Generation"
elif 1925 <= birth_year <= 1945: 
  generation = "The Silent Generation"
elif 1946 <= birth_year <= 1964: 
  generation = "The Baby Boomers"
elif 1965 <= birth_year <= 1979: 
  generation = "Generation X"
elif 1980 <= birth_year <= 1994: 
  generation = "The Millennials"
elif 1995 <= birth_year <= 2012: 
  generation = "Generation Z"
elif 2013 <= birth_year: 
  generation = "Generation Alpha"
else: 
  generation = "no generation"

year_difference = current_year - birth_year
print("You were born", year_difference, "years ago, which means you're part of", generation)
````

This definitely works, but when you look at this big set of `if` statements, it's easy to lose yourself. So, what if we could write something like: 
```
current_year = 2023
birth_year = input("What year were you born? ")
birth_year = int(birth_year)
generation = find_generation()
year_difference = current_year - birth_year
print("You were born", year_difference, "years ago, which means you're part of", generation)
```

Look how much cleaner that is! Well, as you know this is exactly the type of thing that a function can help us with. For example, we could write out script like this:

````python
def find_generation(year): 
  """Function that finds the generation of a given birth year"""
  if 1901 <= year <= 1924:
    generation = "The Greatest Generation"
  elif 1925 <= year <= 1945: 
    generation = "The Silent Generation"
  elif 1946 <= year <= 1964: 
    generation = "The Baby Boomers"
  elif 1965 <= year <= 1979: 
    generation = "Generation X"
  elif 1980 <= year <= 1994: 
    generation = "The Millennials"
  elif 1995 <= year <= 2012: 
    generation = "Generation Z"
  elif 2013 <= year: 
    generation = "Generation Alpha"
  else: 
    generation = "no generation"
  return generation

# Script for
current_year = 2023
birth_year = input("What year were you born? ")
birth_year = int(birth_year)
generation = find_generation(birth_year)
year_difference = current_year - birth_year
print("You were born", year_difference, "years ago, which means you're part of", generation)
````
