### A small hint

By the way, remember how we've been working with numpy? Well, here's an updated version of the script from last lab, but using numpy arrays. 

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(range(25))

# different ms to test
ms = [0, 1, 2, 3, 5, 7, 9]

# constant b
b = -10

for m in ms[::-1]: 
  # calculate the ys with this m
  y_m = m*x + b

  # plot the ys with this m
  plt.plot(x, y_m, label=m)


# Make a legend which labels the m
plt.legend(ms)

# Add a title
plt.title("y = mx + b with different m values and b = -10 (cookie sales)")

# Show the graphic
plt.show()
````
