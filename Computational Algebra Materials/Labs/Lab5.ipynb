# Vary, vary, vary

Now, identify what your `a`, `r`, and `k` are. We will now begin to modify them so we can understand what they do.  This is something you can do with all sorts of equations. 


For example, let's take the equation: 

$$ y = mx + b$$

Lets say we use this equation to represent the money I make froma cookie sale. Here we have: 

$$MoneyMade = (Price)(NumberOfCookies) - (CostOfIngredients)$$

So, the `m` corresponds to the price of each cookie, and the cost of the ingredients is `-b`. 


We want to see what happens as I change `m` and `b`. 



Let's start with changing `m`, the price of each cookie. 

````python
import matplotlib.pyplot as plt
import numpy as np

xs = list(range(25))

# different ms to test
ms = [0, 1, 2, 3, 5, 7, 9]

# constant b
b = -10

for m in ms: 
  y_m = []

  # calculate the ys with this m
  for x in xs:
    y_m.append(m*x + b)

  # plot the ys with this m
  plt.plot(xs, y_m)


# Make a legend which labels the m
plt.legend(ms)

# Add a title
plt.title("y = mx + b with different m values and b = -10 (cookie sales)")

# Show the graphic
plt.show()
````

From this it seems that the higher `m` is, the steeper the line. **This tells me that if I increase the price (`m`), I earn money faster.** I get more money for each sale. 


We could do the same thing changing `b` (ingredient cost), while holding `m` constant:

````python
import matplotlib.pyplot as plt
import numpy as np

xs = list(range(25))

# different bs to test
# negative because we SPEND money to 
# buy ingredients
bs = [0, -10, -20, -30, -50, -70, -90]

# constant m
m = 1

for b in bs: 
  y_m = []

  # calculate the ys with this m
  for x in xs:
    y_m.append(m*x + b)

  # plot the ys with this m
  plt.plot(xs, y_m)


# Make a legend which labels the m
plt.legend(bs)

# Add a title
plt.title("y = mx + b with different b values and m = 1 (cookie sales)")

# Show the graphic
plt.show()
````

From this we can tell that the b value moves the line up and down but doesn't change how steep it is. **This tells me that if I spend more money on ingredients without changing the price, I will ultimately make less money**. In fact, it looks like with $30 or more of ingredient cost I NEVER make any money because my made money never reachers more than 0.
