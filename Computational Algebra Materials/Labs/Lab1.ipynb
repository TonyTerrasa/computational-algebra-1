### Median
Now, let's look at the median. Unlike the mean, we don't have a simple formula for the median. Instead, we have a process for finding the mean which is: 

* Put our data points in order from least to greatest
* Find the point in the middle 

We're in luck, there's also tools to help us with that in Python. For example, if we want to sort the data from lowest to highest, we can use: 
```
baltimore_sorted = sorted(Baltimore)
```

