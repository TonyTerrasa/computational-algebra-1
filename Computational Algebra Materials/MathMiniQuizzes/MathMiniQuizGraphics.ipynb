# Box and Whisker Plots

```python
import matplotlib.pyplot as plt
import numpy as np

times = [10.0, 10.04, 10.04, 10.41, 10.51, 10.51, 10.52, 11.0, 11.01, 11.02, 11.1, 11.2, 12.0, 12.1, 12.2, 12.5, 14]

fig = plt.figure(1, figsize=(9, 4))
ax = fig.add_subplot(111)

ax.boxplot(times,
            vert=False, 
            manage_ticks=True
            )

ax.set_xlabel('seconds')
ax.set_yticks([1])
ax.set_yticklabels([''])

quantiles = np.quantile(times, np.array([0.00, 0.25, 0.50, 0.75, 1.00]))
ax.vlines(quantiles, [0] * quantiles.size, [1] * quantiles.size,
          color='b', ls=':', lw=0.5, zorder=0)
ax.set_ylim(0.5, 1.5)
ax.set_xticks(quantiles)
ax.set_ylim((0.8,1.2))
plt.title('High School Team 100m Times')
plt.show()
```

# Quiz number 2

```python
import numpy as np
import matplotlib.pyplot as plt

ts = np.arange(0, 18.1, 0.5)

rate = 0.5

softball = 10 - rate*ts
baseball = 2 + rate*ts


plt.plot(ts, softball)
plt.plot(ts, baseball)
plt.legend(["Softball Cooler", "Baseball Cooler"])
plt.xlabel("Time in Seconds")
plt.ylabel("Gallons")
plt.xlim((0, 18.1))
plt.xticks(range(19))
plt.title("Water in Softball and Baseball Coolers")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-25, 25)
y1 = x**2 + 4
y2 = np.sin(x / 8)**2

fig, axs = plt.subplots(1, 4)

fig.set_figwidth(25)


axs[0].plot(x, y1)
axs[0].set_title("(a)")

axs[1].plot(y1, x)
axs[1].set_title("(b)")

axs[2].plot(y2, x)
axs[2].set_title("(c)")

axs[3].plot(x, y2*np.exp(x/10))
axs[3].set_title("(d)")

plt.suptitle("Question 1")
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt, patches

T = np.arange(90, 109)
e = np.abs(99-T)


plt.plot(T, e)
plt.plot(T, [5 for i in T], '--')
plt.title("Distance from Normal Body Temperature")
plt.xlabel("Degrees")
plt.ylabel("Degrees from Normal")
plt.ylim(0, 9)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0, 25, 100)
y1 = 2**x1
y2 = (0.5)**x1

fig, axs = plt.subplots(2, 3)

fig.set_figwidth(20)
fig.set_figheight(10)


axs[0, 0].plot(x1, y1)
axs[0, 0].set_title("(a)")

axs[0, 1].plot(x1, y2)
axs[0, 1].set_title("(b)")

axs[0, 2].plot(x1, y2/10000)
axs[0, 2].set_title("(c)")

axs[1, 0].plot(x1, 3*x1 + 4)
axs[1, 0].set_title("(d)")


axs[1, 1].plot(x1, -0.5*x1 + 4)
axs[1, 1].set_title("(e)")


axs[1, 2].plot(x1, y1/100)
axs[1, 2].set_title("(e)")


plt.suptitle("Question 1")
fig.tight_layout()
plt.show()
```
