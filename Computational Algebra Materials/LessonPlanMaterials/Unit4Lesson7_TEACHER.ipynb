# Functions with more than one parameters

One of the cool things about Python is that it allows us to have functions with more than one parameter, like the function below. What do you think the output of the funciton will be?

````python
def add_normal(a, b):
  return a + b

def add_second_half(a, b):
  second_half = b / 2
  return a + second_half
````

````python
print(add_normal(1, 4))
print(add_normal(5, 100))
print(add_normal(10, 3))
````

````python
print(add_second_half(1, 4))
print(add_second_half(5, 100))
print(add_second_half(10, 3))
````

# Practice Time
One a piece of paper, describe the following functions by giving: 
1. Input(s)
2. What happens to the inputs to get the output

````python
def triple(g):
  return 3 * g

def square(x):
  return x * x

def feet_to_meters(feet):
  foot_to_meter_ratio = 0.3048
  return feet * foot_to_meter_ratio

def area(length, width):
  return length * width

def average(a, b, c):
  input_sum = a + b + c
  return input_sum / 3
````

# Answers

```python
def triple(g):
    return 3 * g
```
#### `triple`
- input: one variable, `g`
- output: three times the input `g`. It triples the input

```python
def square(x):
    return x * x
```
#### `square`
- input: one variable, `x`
- output: the input `x` is squared by multiplying it by itself

```python
def feet_to_meters(feet):
    foot_to_meter_ratio = 0.3048
    return feet * foot_to_meter_ratio
```
#### `feet_to_meters`
- input: one variable, `feet`
- output: it seems to convert feet to meters. It takes the distance in `feet` and multiplies it by the ratio of feet to meters to output input distance in meters

```python
def area(length, width):
    return length * width
```
#### `area`
- input: two variables, `length` and `width`
- output: it seems to calculate the area of a rectangle with sides  `length` and `width`. It returns the product of `length` and `width`

```python
def average(a, b, c):
    input_sum = a + b + c
    return input_sum / 3
```
#### `average`
- input: three variables, `a`, `b`, and `c`
- output: it seems to calculate the average of the three given numbers. First it adds the three numbers together, and returns this sum divided by 3

````python
print(triple(5))
print(square(7))
print(feet_to_meters(3))
print(area(6, 8))
print(average(1, 3,4))
````
