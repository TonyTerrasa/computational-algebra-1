# Part 1: Warm-up

You work at a restaurant which sells pizzas and bookies. Write a function called `bill_total` which takes two arguments: `pizzas` and `cookies` (the number of pizzas ordered and the number of cookies respectively) and returns the total cost of the meal. Each pizza is \$5 and each cookie is \$0.50. Don't include any sales tax.

Test your code by running the inputs below to see if you get the correct answer: 

| Pizzas | Cookies | Bill Total (\$) |
| --- | --- | --- |
| 1 | 1 | 5.50 |
| 0 | 0 | 0 |
| 10 | 3 | 51.50 |
| 3 | 104 | 67.00 |

````python
# Warm-up Answer
def bill_total(pizzas, cookies):
  pizza_total = 5 * pizzas # each pizza is $5
  cookie_total = 0.5 * cookies # each cookies is $0.50
  return pizza_total + cookie_total

print("Total 1 pizza and 1 cookie", bill_total(1, 1))
print("Total 0 pizza and 10cookie", bill_total(0, 0))
print("Total 10 pizzas and 3 cookies", bill_total(10, 3))
print("Total 3 pizzas and 104 cookies", bill_total(3, 104))
````

1. Rewrite the code below to use functions to make it less repetitive.

````python
group_1_scores = [14, 30, 18, 24, 29]
group_2_scores = [71, 69, 62, 64, 79]
group_3_scores = [120, 131, 127, 124, 129]


# Your Code Here 
def mean(scores):
  sum_scores = sum(scores)
  N_scores = len(scores)
  return sum_scores / N_scores

def median(scores):
  scores_sorted = sorted(scores)
  median_location = int(len(scores) / 2)
  return scores_sorted[median_location]

# group 1
group_1_average = mean(group_1_scores)
group_1_median = median(group_1_scores)
# group 2
group_2_average = mean(group_2_scores)
group_2_median = median(group_2_scores)
# group 3
group_3_average = mean(group_3_scores)
group_3_median = median(group_3_scores)

print("Group 1 had an average of", group_1_average, "and a median of", group_1_median)
print("Group 2 had an average of", group_2_average, "and a median of", group_2_median)
print("Group 3 had an average of", group_3_average, "and a median of", group_3_median)
````

Now, the area of a regular polygon can be a little more complicated. The formula looks like this: 

$$ A = \frac{1}{2} aP$$

Where: 
- $A$ is the area
- $a$ is the length of the apothem
- and $P$ is the perimeter

The apothem is a line which goes from the center of the polygon to the middle of the the side of the polygon. 

![](https://d138zd1ktt9iqe.cloudfront.net/media/seo_landing_files/shilpi-n-apothem-01-1605076140.png)

2. Make another called `area_reg_polygon` which takes arguments `n_sides` and `side_length` and calculates the area of the regular polygon with `n_sides` sides and a side length of `side_length`. You've been provided a function `apothem` which takes `n_sides` and `side_length` and you can call from within your own function to calculate the apothem. 


You can test your function with code below

````python
from math import tan, pi
def apothem(n_sides, side_length):
  angle = pi / n_sides
  return side_length / (2 * tan(angle))

# Your code goes here
def area_reg_polygon(n_sides, side_length):
  perimeter = n_sides * side_length # students can also use their perimeter function from above
  a = apothem(n_sides, side_length)
  return (1/2) * a * perimeter



# For testing
print("For a regular square (4 sides) with side length 10, the area should be 100. Your answer:", area_reg_polygon(4, 10))
print("For a regular nonagon (9 sides) with side length 6, the area should be 222.54. Your answer:", area_reg_polygon(9, 6))
print("For a regular polygon with 412 sides with side length 1, the perimeter should be 13507.53. Your answer:", area_reg_polygon(412, 1))
````

## Drawing the average rate of change

Recall the the average rate of change between two points of a function is the slope of the connecting line. We can find the slope using the following formula: The average rate of change of $f(x)$ between $x = a$ and $x = b$ is:
$$R = \frac{f(b) - f(a)}{b -a}$$

3. Write a function `average_ROC` which takes parameters `fa`, `fb`, `a`, and `b` and returns the average rate of change between the points $(a, f_a)$ and $(b, f_b)$. 

````python
# Your code here
def average_ROC(fa, fb, a, b): 
  return (fb - fa) / (b - a)


# Test your code by running the following
print(f"Inputs (3, 4) and (4, 5) gave result: {average_ROC(4, 5, 3, 4)} and should be 1")
print(f"Inputs (3, 7) and (8, 5) gave result: {average_ROC(7, 5, 3, 8)} and should be -0.4")
print(f"Inputs (3, 3) and (4, 3) gave result: {average_ROC(3, 3, 4, 3)} and should be 0")
````
