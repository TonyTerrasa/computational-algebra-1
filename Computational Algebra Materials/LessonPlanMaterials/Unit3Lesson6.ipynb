However, we can also make scatter plots using the command `plt.scatter(x_data, y_data)` where the data is treated like a set of x-y pairs as if the data were: 

| x | y |
|---|---|
| `x_data[0]` | `y_data[0]` |
| `x_data[1]` | `y_data[1]` |
| `x_data[2]` | `y_data[2]` |
| ... | ... |


For this reason `x_data` and `y_data` must have the same length.

So, let's look at an example of making a scatter plot: 
```
hours = [5, 5.5, 6, 6.1, 8]
scores = [70, 73, 75, 80, 85]

# import
import matplotlib.pyplot as plt

# make the scatter plot
plt.scatter(hours, scores)

# display the graph
plt.show()
```
