# Part 1: Warm-Up

Your friend takes a video of you dropping a tennis ball from about 4m (12 ft) high. Using Python, you analyze the video and calculate the height of the ball from the ground versus time. The graph is below. 

Using the graph, at what time do you think each of the following happened? Why?
- the ball is dropped
- when the ball hits the ground
- the ball is moving the fastest

(Generation Code is Below)

```python
import numpy as np
import matplotlib.pyplot as plt

h_i = 4
static_time = 1
g = 10

t1 = np.linspace(0, static_time, 100)
t2 = np.linspace(static_time, 5, 1000)

h1 = h_i*np.ones(len(t1))
noise = np.random.normal(size=(len(t1)), scale=0.05)
noise[-3:] = 0
h1 += noise

h2 = -(g/2)*np.square(t2 - static_time) + h_i

t2 = t2[np.where(h2 >= 0)]
h2 = h2[np.where(h2 >= 0)]

bounce_time = t2[-1]
t3 = np.linspace(bounce_time, bounce_time+0.2, 100)
h3 = -(g/2)*np.square(t3 - bounce_time) + (0.3*g * bounce_time)*(t3 - bounce_time)


plt.plot(t1, h1, c='k')
plt.plot(t2, h2, c='k')
plt.plot(t3, h3, c='k')

plt.title("Height of a Dropped Ball")
plt.xlabel("Time (sec)")
plt.ylabel("Meters")
plt.xlim(0, bounce_time+0.2)
plt.ylim(0, 4.3)

plt.xticks([0, 0.5, 1, 1.5, 2])
plt.yticks(range(1,5))


plt.show()
```

# Part 2: Chucking a Ball

Your friend takes another video, this time of you throwing the ball on the football field. Using Python, you analyze the video and calculate the height of the ball and *how far away the ball was from you horizontally*. The graph is below. 

Using the graph, determine:
- where on the graph the ball is **thrown**
- how far the ball traveled before hitting the ground
- about how many times it bounced
- how far the ball was when it began to roll
- how far the ball was thrown
- how far away the ball was when it reached its highest point 


(Generation Code is Below)

```python
import numpy as np
import matplotlib.pyplot as plt

g = 10
rebound_factor_y = 0.5
rebound_factor_x = 0.98

v_x = 10
v_0y = 15
y_0 = 1.6
x_0 = 0

t = np.linspace(0, 10, 10000)
x = np.zeros(len(t))
y = np.zeros(len(t))

t_bounce = 0

for i, t_ in enumerate(t): 
  x_ = v_x * (t_ - t_bounce) + x_0
  y_ = (-g/2)*(t_ - t_bounce)**2 + v_0y * (t_ - t_bounce) + y_0
   

  x[i] = x_
  y[i] = y_

  # bounce detected
  if y_ <= 0:
    y_0 = 0
    v_0y = -rebound_factor_y * ( -g*(t_ - t_bounce) + v_0y )

    x_0 = x_
    v_x *= rebound_factor_x

    t_bounce = t_



plt.plot(x, y, c='k')


plt.title("A Thrown Ball")
plt.xlabel("Horizontal Distance from Thrower (m)")
plt.ylabel("Height (m)")
plt.xlim(0, x[-1]+4)
plt.ylim(0, 15)

#plt.xticks([0, 0.5, 1, 1.5, 2])
#plt.yticks(range(1,5))


plt.show()
```

# Pollutant AQI

Wildfires can be dangerous for a lot of reasons. One of the reasons is that they emit harmful air pollutants into the air, for example PM2.5, fine particles that can negatively impact lunch functioning. We want to understand how the pollution spreads in the surrounding environment from the center of the fire (not necessarily the edge). Governments often release a metric called the AQI (air quality index), a way of measuring the concentration of a pollutant in the air. The higher the AQI, the more of the pollutant there is and the more dangerous it is. Below is a chart of how harmful pollutants at each AQI are.

# The Parabolic Mirror


Did you know that the Olympic torch is traditionally lit with a mirror? How is that possible? Well, its not just any mirror, its a parabolic mirror, shaped like a parabola. See the pictures below. 


Parabolas have an important property called the focal point, and while exactly how this works is beyond the scope of our class, we can think of the focal point as the place where the incident light is most concentrated. So, if the sunlight enters the parabolic mirror, the focal point is where we will most feel the heat of the sun when the light is reflected back. If we locate our unlit torch at the focal point (F in the diagram above) of the mirror, the high intensity light will heat it up until it lights! Understanding the  focal point of mirrors and lenses is part of the study of optics. 

```python

import numpy as np
import matplotlib.pyplot as plt


r = 3
x = np.linspace(0, 2*r, 100)
y = (1/4)*np.square(x - r) + 0.5

plt.plot(x, y, c='k')

# formatting
plt.ylim(0, 3)

plt.yticks(np.arange(0, 3+0.25, 0.5))
plt.xticks(np.arange(0, 2*r+1, 1))


plt.title("Shape of a Parabolic Mirror")
plt.xlabel("x (ft)")
plt.ylabel("y (ft)")

plt.show()
```
