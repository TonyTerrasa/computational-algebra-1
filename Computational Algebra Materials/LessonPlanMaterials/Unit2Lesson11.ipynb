# Part 1: While Loops

While loops, much like for loops can be used to execute code repeatedly. Check out the explanation and example from W3Schools<sup>1</sup>

<blockquote>
With the <code>while</code> loop we can execute a set of statements as long as a condition is true.


Print i as long as i is less than 6: 
<pre>
<code>i = 1
while i < 6:
  print(i)
  i += 1</code>
</pre>

Note: remember to increment i, or else the loop will continue forever.

The while loop requires relevant variables to be ready, in this example we need to define an indexing variable, i, which we set to 1.
</blockquote>

In this case the **condition** is `i < 6`. So, as long as `i < 6` is `True`, the coded will run. As soon as `i < 6` becomes `False`, it will stop and move on with the code. Now, run the code in the example to see the output

````python
i = 1
while i < 6:
  print(i)
  i += 1
````
