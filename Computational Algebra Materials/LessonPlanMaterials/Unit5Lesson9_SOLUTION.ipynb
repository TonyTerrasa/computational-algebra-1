````python
from numpy import linspace

# First, adjust the functions so 
# we don't have the 000 at the end pf
# each y value
def p(y):
  return 0.720 * 12 * y

def l(y):
  return 25*(1.08)**y

def apply_function(f, x):
  """
  Given a function f and a list x
  Return a new list with values: 
  [f(x[0]), f(x[1]), ...]
  """
  return [f(x_i) for x_i in x]


def find_intercept(f, g, a, b):
  """
  Given two functions: f and g
  and an interval a < x < b
  Attempts to find an intersection of f and g
  On the interval. Returns the x value where 
  they intersect
  """

  x = linspace(a, b, 1000)

  for x_, f_, g_ in zip(x, apply_function(f, x), apply_function(g, x)):
    if abs(f_ - g_) < abs(f_*0.01):
      return x_
  else:
    print("Intercept failed. Did not find an intercept. Check the range or make it smaller.")


# Your code here

# showing the function from years 0 to 6
y = linspace(0, 6, 1000)

# calculate the values of p and l on this range
p_values = apply_function(p, y)
l_values = apply_function(l, y)


# your code here

# the time when we finish paying the loan
# we can see from the graph that the two
# functions intersect between 3.5 and 4.5 
# years into the loan
y_finish = find_intercept(p, l, 3.5, 4.5)

import matplotlib.pyplot as plt

plt.plot(y, l_values, label="Total Loan Value", c='k')
plt.plot(y, p_values, label="Total Amount Payed", c='grey')
plt.plot([y_finish, y_finish], [0, p(y_finish)], linestyle="--", c='g')
plt.legend()

plt.title("A \$25K Car with Monthly Payments of \$720")
plt.ylabel("Thousands of $")
plt.xlabel("Years from Purchase")
plt.text(4.01,25,"Loan finished being\npayed after %.2f years" % y_finish)
plt.text(4.01,20,"Total amount payed: \n$%.2fK" % p(y_finish))

plt.xlim(0, 6)
plt.ylim(0, 50)

plt.xticks(
    range(7),
    ["Purchase"] + list(range(1,7))
)

plt.yticks(
    [10, 20, 25, 30, 40, 50],
)



plt.show()
````
