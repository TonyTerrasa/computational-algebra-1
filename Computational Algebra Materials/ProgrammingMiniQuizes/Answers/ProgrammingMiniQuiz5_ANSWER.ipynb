## Calculations on a Circle (~ Reprise ~)

Last quiz, we looked at the formulas for area and circumference of a circle based on the radius. 

<blockquote>
<em>
The area of a circle can be calculated be the simple formula: 

$Area = \pi \cdot radius \cdot radius $

The circumference however can be calculated using the following formula: 

$Circumference = 2 \cdot \pi \cdot radius $

![](https://www.dummies.com/wp-content/uploads/439621.image0.jpg)

Image from dummies.com<sup>1</sup>
</em>
</blockquote>

This we will explore the results. Below is one way to implement these functions. You may use them in the rest of this quiz.


````python
from math import pi
def area(radius):
  return pi * radius**2

def circumference(radius):
  return 2 * pi * radius
````

## Part 3: Inverting the Function


Let's go back to the circles. What happens if you are given the area of a circle and you want to find the radius? There is an equation for that! See the equation below. 

$Radius = \sqrt{\frac{Area}{\pi}}$

Complete the Python code below to make a function that. You will need to import the `sqrt` function and the number `pi` from the `math` package. 

Once imported, `sqrt` can be used like this:


```
sqrt(number)
```

Similarly, once imported, `pi` can be used like any other variable:
```
a = 10 + pi
```


````python
# Import
from math import sqrt, pi

# Function
def radius_from_area(area):
  return sqrt(area/pi)
````

````python
# Use this code to test your function
print(f"For area 6 the radius is 1.38 and your answer was {str(round(radius_from_area(6), 2))}")
print(f"For area 144 the radius is 6.77 and your answer was {str(round(radius_from_area(144), 2))}")
print(f"For area 57 the radius is 4.26 and your answer was {str(round(radius_from_area(57), 2))}")
````
