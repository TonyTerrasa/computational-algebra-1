## Calculations on a Circle

The area of a circle can be calculated be the simple formula: 

$Area = \pi \cdot radius \cdot radius $

The circumference however can be calculated using the following formula: 

$Circumference = 2 \cdot \pi \cdot radius $

## Write your functions

Write a function, which takes one parameter - the radius - and returns the **area** of the circle.

```python
pi = 3.14
def area(radius):
  # written without using exponents, but could also be 
  # written as (pi * radius**2)
  return pi * radius * radius
```

Write a function, which takes one parameter - the radius - and returns the **circumference** of the circle.

```python
pi = 3.14
def circumference(radius):
  return 2 * pi * radius
```

## Calling your functions

1. Using your functions, calculate the area and circumference of a circle with a radius of `6`. Print your answer using label text.

```python
area_6 = area(6)
circumference_6 = circumference(6)
print("For a circle of radius 6, the area is", area_6, "and the circumference is", circumference_6 )
```

2. Using your functions, calculate the area and circumference of a circle with a radius of `11`. Print your answer using label text.

```python
area_11 = area(11)
circumference_11 = circumference(11)
print("For a circle of radius 11, the area is", area_11, "and the circumference is", circumference_11)
```
